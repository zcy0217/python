# 第六題

請製作一個程式 設定 範圍(head) 跟 目標(k) 完成以下輸出 輸入

EX:1



    Input: head = [1,2,3,4,5], k = 2
    Output: [2,1,4,3,5]
    
EX:2


    Input: head = [1,2,3,4,5], k = 3
    Output: [3,2,1,4,5]
    
EX:3
    
    Input: head = [1,2,3], k = 5
    Output: [1,2,3]


```python
def reverseKGroup(head,k,ans = []):
    if len(head)<k:
        return ans+head
    else:
        reverse_num = []
        for i in range(k):
            reverse_num.append(head[0])
            head.pop(0)
        reverse_num = reverse_num[::-1]
        return reverseKGroup(head,k,ans+reverse_num)

reverseKGroup([1,2,3],5)
```
