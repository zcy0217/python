信用卡號碼驗證

信用卡號碼遵循特定形式。信用卡號碼必須含有13~16位數。起始位數必須是： 
 4代表Visa cards 
 5代表Master cards 
 37代表American Express cards 
 6代表Discover cards 
1954年，IBM的Hans Luhn提出一個驗證信用卡號碼的演算法。此演算法用在判斷輸入的卡號是否正確，或在信用卡是否正確被掃描上很有用。信用卡號遵循此來產生有效驗證，這種作法稱為Luhn check或Mod 10 check，說明如下(為了方便解釋，這裡以卡號4388576018402626為例)： 


1. 由左到右，每隔一位數便取其兩倍數字。如果該位數的兩倍數字為兩位數，便將該兩位數相加，以取單一位數數字。
     請參考附件圖 
2. 接著將步驟1所有的單位數數字相加。
4+4+8+2+3+1+7+8=37 
3. 再由右到左，將出現於卡號奇數位置的所有位數相加。
6+6+0+8+0+7+8+3=38 
4. 將步驟2與步驟3的結果相加。
37+38=75

5. 如果步驟4的結果能夠被10整除，則卡號便是有效的；反之，則是無效的。
比方說，卡號4388576018402626是無效的，但卡號4388576018410707是有效的。

```python
print("\n請注意!信用卡卡號為整數")
print("輸入錯誤三次將會跳出")
time = 0 #設定輸入次數
while(time<3):
    code = list(input("請輸入您的信用卡卡號:"))    #輸入號碼
    for i in range(0, len(code)):    
        code[i] = int(code[i])       #讓輸入的號碼轉為int的list
    if 13 <= len(code) <= 16 :       #限定輸入字數
        if code[0] == 4 or code[0] == 5 or code[0] == 6 or code[:2] == [3,7]:#限定開頭數字
            first_num = 0
            for n in range(0,len(code),2): #隔位在code中取元素
                if int(code[n])*2 <10:     #每位數取其兩倍小於十時
                    first_num+=int(code[n])*2 #每位數取其兩倍數字相加
                else:
                    a=int(code[n])*2
                    b=a%10+a//10      # *2大於十的時候 兩位數相加
                    first_num+=b
            two_code = code[::-1]   #反轉list中順序
            two_num = 0
            for z in range(0,len(two_code),2):#卡號奇數位置的所有位數相加
                 two_num+= int(two_code[z])
            total = two_num+first_num
            if total%10 == 0: #總數被10整除
                print("real")
                break
            else:
                print("false")
                time+=1
                print(f"輸入錯誤{time}次")
                continue
        else:
            time+=1
            print("開頭輸入錯誤")
            print(f"輸入錯誤{time}")
    else:
        print("輸入號碼限制")
        time+=1
if time == 3:
    print("輸入錯誤超過三次，跳出")
else:
    print("謝謝您的使用")
```
