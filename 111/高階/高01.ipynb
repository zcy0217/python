```python
'''

請製作一個可以將矩陣向右轉90度的程式

範例輸入:
[[1,2,3],[4,5,6],[7,8,9]]

範例輸出:
[[7,4,1],[8,5,2],[9,6,3]]

'''
```

```python
def rotate(matrix):
    matrix.reverse()
    for i in range(0,len(matrix)):
        for j in range(i+1,len(matrix)):
            matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]
    return matrix
rotate([[1,2,3],[4,5,6],[7,8,9]])
```
